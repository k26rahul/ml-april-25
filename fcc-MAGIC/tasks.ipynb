```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
array = np.array([10, 20, 30, 40, 50, 50])
n = len(array)
```

```python
mean = sum(array) / n
mean
```

```python
if n % 2 == 1:
  median = array[n//2]
else:
  median = (array[n//2] + array[n//2-1]) / 2

median
```

```python
freq = {}
for num in array:
  freq[num] = freq.get(num, 0) + 1

mode = max(list(freq.keys()))
mode
```

**Markers**: `'o'` (○), `'v'` (▼), `'^'` (▲), `'<'` (◁), `'>'` (▷), `'s'` (■), `'*'` (✶), `'+'` (＋), `'x'` (✕), `'D'` (◆)  

**Lines**: `'-'` (solid), `'--'` (dashed), `'-.`' (dash-dot), `':'` (dotted)

```python
def plot_stats():
  plt.figure(figsize=(8, 4))

  plt.scatter(range(len(array)), array)
  plt.axhline(mean, color='r', label=f'{mean=:.2f}')
  plt.axhline(median, color='g', label=f'{median=:.2f}')
  plt.axhline(mode, color='b', label=f'{mode=:.2f}')

  plt.legend()


plot_stats()
```
