```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ages = [15, 20, 45, 30, 40, 25, 35, 50, 1000, 1000]
ages = [15, 20, 20, 30, 40, 25, 35, 50]
```

```python
mean = np.mean(ages)
median = np.median(ages)
mode = np.argmax(np.bincount(ages))
var = np.var(ages)
std = np.std(ages)

print(f'{mean=:.2f}')
print(f'{median=:.2f}')
print(f'{mode=:.2f}')
print(f'{var=:.2f}')
print(f'{std=:.2f}')
```

```python
def plot_stats():
  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

  ax1.scatter(range(len(ages)), ages, label='values')
  ax1.axhline(mean, color='r', label='mean')
  ax1.axhline(median, color='g', linestyle='--', label='median')
  ax1.axhline(mode, color='y', linestyle='--', label='mode')
  ax1.legend()

  ax2.scatter(range(len(ages)), ages, label='values')
  ax2.axhline(mean, color='r', label='mean')
  ax2.axhline(mean + std, color='r', linestyle='--', label='mean + std')
  ax2.axhline(mean - std, color='r', linestyle='--', label='mean - std')
  ax2.legend()


plot_stats()
```

## Sum Squared Residuals - shows total prediction error

Residual = actual value minus predicted value  
Prediction = whatever we predicted  
Sum Squared Residuals = how wrong we were (add all squared residuals)  
Mean Squared Error (MSE) = average of squared residuals

```python
ages_actual = np.array([15, 20, 30])
ages_predicted = np.array([12, 18, 35])
```

```python
residuals = ages_actual - ages_predicted
squared_residuals = residuals ** 2
sum_squared_residuals = sum(squared_residuals)
mean_squared_error = np.mean(squared_residuals)
root_mean_squared_error = mean_squared_error ** 0.5

print(f'residuals: {residuals}')
print(f'squared_residuals: {squared_residuals}')
print(f'{sum_squared_residuals=:.2f}')
print(f'{mean_squared_error=:.2f}')
print(f'{root_mean_squared_error=:.2f}')
```

```python
def plot_residuals():
  plt.scatter(range(len(ages_actual)), ages_actual, color='b', label='ages_actual')
  plt.scatter(range(len(ages_predicted)), ages_predicted, color='y', label='ages_predicted')

  for i in range(len(ages_actual)):
    plt.plot([i, i], [ages_actual[i], ages_predicted[i]], color='orange', linestyle='--')

  plt.legend()


plot_residuals()
```
